You realize that the disasters from "Tsunami" to the end of the list are forms of water damage.

Programmatically create a new list called water_incidents that includes all the disasters from "Tsunami" to "Tidal Surge".

Do NOT manually count to figure out the index position of "Tsunami".
Do NOT manually type out a new list.
