```python
df.set_index('Year', inplace=True)
```

```python
columns_to_plot = ["Avg. cable bill", "Median household income", "Social Security cost-of-living adjustment"]
```

```python
df[columns_to_plot].plot()
```

```python
old_bill = df["Avg. cable bill"].iat[0]
old_bill
```

```python
df["Avg. cable bill"] = \
((df["Avg. cable bill"] - old_bill) / old_bill) * 100

df
```

```python
## iat for fixed old income
old_income = df["Median household income"].iat[0]
old_income
```

```python
df["Median household income"] = pd.to_numeric(df["Median household income"], errors='coerce')
df["income_chg"] = ((df["Median household income"] - old_income) / old_income) * 100
```

```python
df["income_chg"] = \
((df["Median household income"] - old_income) / old_income) * 100

df
```

```python
 
df.plot(x = "Year",
       y = ["bill_chg", "income_chg"])
```
